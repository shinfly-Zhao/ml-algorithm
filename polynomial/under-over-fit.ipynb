### 过拟合 欠拟合

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(666)
x = np.random.uniform(-3.,3,size=100)
X = x.reshape(-1,1)
```

```python
y = 0.5 * x**2 +x +2 +np.random.normal(0,1,100)
```

```python
plt.scatter(x,y)
```

```python
# 使用均方误差衡量模型 
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
lin_reg = LinearRegression()
lin_reg.fit(X,y)
lin_predict = lin_reg.predict(X)
```

```python
# 均方误差
mean_squared_error(y,lin_predict)
```

```python
# 多项式回归
def PoplynominalRegression(degree):
    return Pipeline([
        ("poly",PolynomialFeatures(degree=degree)),
        ("stand",StandardScaler()),
        ("lin_regression",LinearRegression())
    ])
```

```python
poly_regression2 = PoplynominalRegression(degree=2)
poly_regression2.fit(X,y)
poly_predict = poly_regression2.predict(X)
```

```python
# 均方误差
mean_squared_error(poly_predict,y)
```

```python
plt.scatter(x,y)
plt.plot(np.sort(x),poly_predict[np.argsort(x)],"r")
```

```python
poly_regression2 = PoplynominalRegression(degree=10)
poly_regression2.fit(X,y)
poly_predict2 = poly_regression2.predict(X)
mean_squared_error(poly_predict2,y)
plt.scatter(x,y)
plt.plot(np.sort(x),poly_predict2[np.argsort(x)],"r")
```

```python
poly_regression3 = PoplynominalRegression(degree=200)
poly_regression3.fit(X,y)
poly_predict3 = poly_regression3.predict(X)
mean_squared_error(poly_predict2,y)
plt.scatter(x,y)
plt.plot(np.sort(x),poly_predict3[np.argsort(x)],"r")
```
