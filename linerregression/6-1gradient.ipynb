梯度下降法模拟

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
plot_x = np.linspace(-1,6,141)  # 随机生成141个随机数
plot_y = (plot_x - 2.5) ** 2 -1  # 模拟损失函数
```

```python
plt.plot(plot_x,plot_y)
plt.show()
```

```python
# 梯度下降法求最优解
def dJ(theta):
    return 2*(theta-2.5)
# 计算损失函数的结果值
def J(theta):
    return (theta - 2.5) ** 2 -1
```

```python
theta = 0.0 # 初始值
eta = 0.1 # 学习率
epsilon = 1e-8
while True:
    gradient = dJ(theta)
    last_theta = theta
    theta = theta - eta*gradient
    if (abs(J(theta) - J(last_theta))< epsilon):
        break
print(theta)
print(J(theta))
    
```

```python
thetaList = []
def gradient_descent(initial_theta,eta,epsilon = 1e-8):
    """
    theta # 初始值
    eta   # 学习率
    """
    theta = initial_theta
    thetaList = [theta]
    epsilon = 1e-8
    while True:
        gradient = dJ(theta)
        last_theta = theta
        thetaList.append(theta)
        theta = theta - eta*gradient
        if (abs(J(theta) - J(last_theta))< epsilon):
            return thetaList
            
def plot_theta_history(thetaList):
    # 绘制过程 
    plt.plot(plot_x,J(plot_x))
    plt.plot(np.array(thetaList),J(np.array(thetaList)),color="red",marker="+")
    plt.show()
```

```python
thetaList = gradient_descent(initial_theta=0.01,eta=0.01)
plot_theta_history(thetaList)
```

```python
thetaList = gradient_descent(initial_theta=0.01,eta=0.8)
plot_theta_history(thetaList)
```
