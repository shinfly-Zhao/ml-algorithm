### 使用梯度上升法求解主成分

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = np.empty((100,2))
X[:,0] = np.random.uniform(0,100,size=100)
X[:,1] = 0.75 * X[:,0]+3 + np.random.uniform(0,10,size=100)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# 样例均值归0  样本的空间分布是不变的
def demean(X):
    return X - np.mean(X,axis=0)
```

```python
a = demean(X)
np.sum(a[:,0])
np.sum(a[:,1])
```

```python
plt.scatter(a[:,0],a[:,1])
```

### 梯度上升法

```python
def f(w,X):
    return np.sum((X.dot(w)**2)) /len(X)
```

```python
def df_max(w,X):
    return X.T.dot(X.dot(w)) *2 / len(X)
```

```python
def direction(w):
    # 始终是单位f方向向量 只代表方向
    return w / np.linalg.norm(w)  # 模
def gradient_accent(df,X,initial_w,eta,n_iters=1e4,espilion = 1e-8):
    w = direction(initial_w)
    cur_iter = 0
    while cur_iter < n_iters:
        
        gradient = df(w,X)
        lase_w = w
        w = w +eta * gradient
        w = direction(w)  # 注意
        if (abs(f(w,X) - f(lase_w,X)) < espilion):
            break
        cur_iter+=1
    return w
```

```python
initial_w = np.random.random(X.shape[1])  # 不能从0开始
eta = 0.01
# 不能进行数据标准化
```

```python
w = gradient_accent(df_max,a,initial_w,eta=eta)
```

```python
plt.scatter(a[:,0],a[:,1])
plt.plot([0,w[0]*30],[0,w[1]*30],color='r')
```

```python
# 求解第n个主成分
X = a
X2 = np.empty(X.shape)
for i in range(len(X)):
    # 去除在第一分量上的成分
    X2[i] = X[i] - X[i].dot(w) * w
```

```python
plt.scatter(X2[:, 0], X2[:,1])
```

```python
initial_w = np.random.random(X2.shape[1])  # 不能从0开始
eta = 0.01
w = gradient_accent(df_max,X2,initial_w,eta=eta)
plt.scatter(X2[:,0],X2[:,1])
plt.plot([0,w[0]*30],[0,w[1]*30],color='r')
```
