### PCA降噪

```python
from sklearn import datasets
import numpy as np 
digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
noise_digits = X + np.random.normal(0,4,size=X.shape)
noise_digits.shape
```

```python
example_digits = noise_digits[y==1,:][:10]
example_digits[1]
```

```python
for num in range(1,10):
    X_num = noise_digits[y==num,:][:10]
    example_digits = np.vstack((example_digits,X_num))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(example_digits[1],example_digits[1])
```

```python
fig, ax = plt.subplots(
    nrows=2,
    ncols=5,
    sharex=True,
    sharey=True, )

ax = ax.flatten()
for i in range(10):
    img = example_digits[y==1,:][1].reshape(8, 8)
    ax.imshow(img, cmap='binary', interpolation='nearest')

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
def plot_digits(data):
    fig,axes = plt.subplots(10,10,figsize=(10,10),subplot_kw={"xticks":[],"yticks":[]},gridspec_kw = dict(hspace=0.1,wspace=0.1))
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap='binary', interpolation='nearest')
    plt.show()
```

```python
plot_digits(example_digits)
```
