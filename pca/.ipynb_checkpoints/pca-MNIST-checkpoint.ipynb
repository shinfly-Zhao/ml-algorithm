### MNIST手写数据

```python
import numpy as np 
from sklearn.datasets import fetch_mldata
import matplotlib.pyplot as plt
```

```python
mnist = fetch_mldata(dataname="MNIST original",data_home="./")
```

```python
X = mnist["data"]
y = mnist["target"]
```

```python
X_train = np.array(X[:60000],dtype=float)
y_train = np.array(y[0:60000],dtype=float)
X_test = np.array(X[60000:],dtype = float)
y_test = np.array(y[60000:],dtype = float)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X_train)
X_train_reduction = pca.transform(X_train)
```

```python
X_train_reduction
```

```python
for i in range(10):
    plt.scatter(X_train_reduction[:100][y_train==i,0],X_train_reduction[:100][y_train==i,1])
plt.show()
```
